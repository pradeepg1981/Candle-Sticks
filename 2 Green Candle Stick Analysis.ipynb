```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dts
import scipy.stats as stats
import yfinance as yfin

import requests
import xlsxwriter
from datetime import date
```

```python
def percDiff(a,b):
    perc= ((b-a)*100)/a
    return perc
```

```python
def stockName(fileName):
    data=pd.read_csv(fileName)
    stocks=data['Symbol']
    for i in range(0,len(stocks)):
        stocks[i]=stocks[i]+'.NS'
    return stocks
```

```python
def nifty_delta(fileName):
    stocks=stockName(fileName)
    # stocks=['INFY.NS','IBREALEST.NS']
    TPE=[] #trailing PE
    CMP=[] #current market price
    SECTOR=[]
    fiftyMovAvg=[]
    delta=[]
    two_CS=[]
    perc=0
    dict_delta={}
    for i in range(0,len(stocks)):
        perc=0
        trend=[]
        ticker=yfin.Ticker(stocks[i])
        lst=ticker.info.keys()
        data=ticker.history(start='2021-03-08',end='2021-03-16')
        CMP.append(ticker.info['regularMarketPrice'])
        SECTOR.append(ticker.info['sector'])
        if (list(lst).count('fiftyDayAverage')!=0):
            fiftyMovAvg.append(ticker.info['fiftyDayAverage'])
        else:
            fiftyMovAvg.append('NaN')
        if (list(lst).count('trailingPE')!=0):
            TPE.append(ticker.info['trailingPE'])
        else:
            TPE.append('NaN')

        for l in range(0,len(data)):
            perc+=percDiff(data['Open'][l],data['Close'][l])
            trend.append(perc)
        delta.append(perc)
        
        iCnt=0 ## to count the continuous UP trends, 1->2 green candle sticks
        for k in range(0,len(data)):
            if(k+1<len(data)):
                if(trend[k]>0 and trend[k+1]>0):
                    iCnt+=1
        if iCnt>1:
            two_CS.append('Yes')
        else:
            two_CS.append('No')
    dict_delta={'Stocks':stocks,'Trailing PE':TPE,'Delta':delta,'CMP':CMP,'50DMA':fiftyMovAvg,'Sector':SECTOR,'2 Green CS':two_CS}
    return dict_delta
```

```python
dict_delta=nifty_delta('NIFTY50.csv')
df_nifty50=pd.DataFrame(dict_delta)
```
